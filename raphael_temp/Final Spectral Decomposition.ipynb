```python
import tensorflow as tf 
import keras as ks 
import numpy as np
import pandas as pd
import script_config as sc

#Location of the data_folder 
data_folder      ="/nvme/drive_1/NTDS_Final/"
users            = pd.read_csv(data_folder+"usersdata.csv",delimiter='\t',header=None).values
filtered_users   = pd.read_csv(data_folder+"filtered_users.csv",delimiter=',').values
```

```python
def create_spectral_decomposition(adjacency): 
    
    n_nodes              = tf.shape(adjacency)[0]
    
    degrees_inv_sqrt     = tf.reshape(
                            tf.reciprocal(
                              tf.sqrt(
                                tf.cast(
                                    tf.reduce_sum(adjacency,axis=0),tf.float32))), 
                                     [n_nodes,1])
    
    normalized_laplacian = tf.subtract(tf.eye(n_nodes),
                            tf.multiply(tf.cast(adjacency,tf.float32),
                              tf.matmul(degrees_inv_sqrt, 
                              tf.cast(tf.transpose(degrees_inv_sqrt),tf.float32) )))

    e_values, e_vectors   = tf.linalg.eigh(normalized_laplacian)
    
    features              = tf.multiply(e_values,e_vectors[0,:])
    
    return features

for i in range(len(filtered_users)):
    adj              = np.load(data_folder+"local/adjacency_"+str(i)+".npy")
    tf_adj           = tf.convert_to_tensor(adj,dtype=tf.float32)
    features         = create_spectral_decomposition(tf_adj)
    with tf.Session() as sess:
        sess.run(tf_adj)
    print(i)
    
```
